#### Eastcom CRS Global Marketing Distribution

```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB


from pyecharts import options as opts 
from pyecharts.charts import Geo, Grid, Bar
from pyecharts.globals import GeoType,ThemeType,SymbolType #Geo图的类型 主题 涟漪图形符号

data=[ ("Cuba",700),("Hong Kong",1500),("Iran",2000),
      ("Egypt",300),('Saudi Arabia',100),
      ]
data = sorted(data,key=lambda x:x[1],reverse=True) #倒序排列
data_x = [x[0] for x in data]
data_y = [x[1] for x in data]

geo = Geo().add_schema(maptype='world',center= [21.824312,39.074208])#以希腊为中心
#增加单个的经纬度
#geo.add_coordinate(name="China",longitude=104.195,latitude=35.675)
#geo.add_coordinate(name="Australia",longitude=133.775,latitude=-25.274)

#以Json文件形式增加世界国家的经纬度
geo.add_coordinate_json('D:/学习/数据分析/数据集/目标世界国家GEO.json')


Map_geo = (
    geo.add("",data,type_=GeoType.EFFECT_SCATTER,
         symbol_size=15,
         # label_opts = opts.LabelOpts(is_show=True,formatter= '{a}',), 
         #  effect_opts=opts.EffectOpts(is_show=True,color='#f948f7') 
       ) #热度效果HEATMAP
    .add("",
         [("China","Hong Kong"),("China","Cuba"),("China","Egypt"),("China","Iran"),('China','Saudi Arabic')],
         type_=GeoType.LINES, 
         effect_opts=opts.EffectOpts(symbol=SymbolType.ARROW,symbol_size=6,color='#5f99bb'),
         linestyle_opts=opts.LineStyleOpts(curve=0.2,color="#B0E2FF"))
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True,formatter= '{b}',)) # 国家名标签
    .set_global_opts(title_opts=opts.TitleOpts(title="Eastcom CRS Global Marketing Distribution",
                                              subtitle="Upto Year 2021",pos_right="center",pos_top="5%"), 
                     visualmap_opts=opts.VisualMapOpts(max_=max(data_y),min_=min(data_y),pos_top="8%",
                                                       is_piecewise=True))

)

bar = (
        Bar()
        .add_xaxis(xaxis_data=data_x)
        .add_yaxis(
            series_name="",
            y_axis=data_y,
           
            category_gap='20%',
            
            label_opts=opts.LabelOpts(
                is_show=True, position="right", formatter="{b} : {c}"
            ),
        )
        .reversal_axis()
        .set_global_opts(
            xaxis_opts=opts.AxisOpts(
                max_=max(data_x), axislabel_opts=opts.LabelOpts(is_show=False)
            ),
            yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(is_show=False)),
            tooltip_opts=opts.TooltipOpts(is_show=False),
            visualmap_opts=opts.VisualMapOpts(
                is_calculable=True,
                dimension=0,
                pos_left="10",
                pos_top="top",
                range_text=["High", "Low"],
                range_color=["lightskyblue", "yellow", "orangered"],
                textstyle_opts=opts.TextStyleOpts(color="#ddd"),
                min_=min(data_y),
                max_=max(data_y),
            ),
        )
    )
grid_chart = (
        Grid()
        .add(
            bar,
            grid_opts=opts.GridOpts(z=7,
                pos_left="10", pos_right="80%", 
                pos_top="70%", pos_bottom="5",
                
            ),
        )      
        .add(Map_geo, grid_opts=opts.GridOpts(z=5,))
    )

grid_chart.load_javascript()
```
